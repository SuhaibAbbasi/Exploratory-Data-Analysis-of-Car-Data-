#  **Exploratory Data Analysis (EDA) of Car Data** 






## Ouput Should Look Like This :



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Automobiles = pd.read_csv("automobiles.csv")
Automobiles.head()
```

```python
Automobiles.info()
```

```python
g = sns.relplot(data = Automobiles,
                x = "horsepower",
                y = "price", hue = "fuel-type",
                palette = "dark",
                size = "engine-size",
                sizes = (1,200),
                style = "num-of-cylinders",
                col = "num-of-doors")

sns.set_style('darkgrid')
sns.set_context('notebook')
g.fig.set_figheight(5)
g.fig.set_figwidth(10)

plt.show()
```

# Exploratory Data Analysis

```python
g = sns.relplot(data = Automobiles,
                x = "horsepower",
                y = "price", hue = "num-of-cylinders",
                palette = "dark",
                size = "engine-size",
                sizes = (1,200),
                style = "aspiration",
                col = "num-of-doors")

sns.set_style('darkgrid')
sns.set_context('notebook')
g.fig.set_figheight(5)
g.fig.set_figwidth(10)

plt.show()
```

# Result

### It's evident that price and horsepower exhibit a significant correlation across the majority of vehicles, though there are notable exceptions. Cars featuring eight-cylinder engines and two doors typically boast the highest price tags. Moreover, cars powered by diesel engines with turbo aspiration tend to command higher prices. Nonetheless, there exists a peculiar case where a car with high horsepower is priced lower, fueled by gasoline instead. Further examination suggests that this anomaly stems from its larger engine size and turbo aspiration. Additionally, curb weight emerges as a contributing factor, with wheelbase exerting a slight influence on horsepower increase.

```python
# Selecting only numeric columns for correlation calculation
numeric_columns = Automobiles.select_dtypes(include=np.number)

# Calculating correlation with price and sorting values in descending order
correlation = numeric_columns.corr()["price"].sort_values(ascending=False)

# Printing the correlation results
print(correlation)
```
